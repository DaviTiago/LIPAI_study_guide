```python
for column in df.columns:    
        print(f"Coluna: {column} ")
        print(f"  Média: {df[column].mean(): 2f}")
        print(f"  Mediana: {df[column].median(): 2f}")
        print(f"  Desvio padrão: {df[column].std(): 2f}")
        print(f"  Mínimo: {df[column].min(): 2f}")
        print(f"  Máximo: {df[column].max(): 2f}")
        print(f"\n {'-'*40}\n")
```

2) Use o dataset escolhido e mostre por meio das representações gráficas (histogramas) as principais características. Procure usar as funções apresentadas nesta atividade.

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Class', data=df)
plt.title('Distribution of Wine Classes')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
# plot Qantidade de amostras por classe de vinho
```

```python
classe = (df['Class']).unique()
    
for column in df.columns[1:]:
    fig, axs = plt.subplots(1, 3, figsize =(15, 5), sharey=True)
    fig.suptitle(f'Histogram of {column} by Class', fontsize=16)

    for i, cls in enumerate(classe):
        sns.histplot(df[df['Class'] == cls][column], bins=15, kde=True, ax=axs[i])
        axs[i].set_title(f'Class {cls}')
        axs[i].set_xlabel(column)
        axs[i].set_ylabel('Frequency')

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()


  
```

3) O dataset Iris não possui valores ausentes. A base escolhida também pode conter todos os descritores por cada amostra. Use um dataset e sobre as amostras impute valores ausentes usando a média, mediana ou outras medidas.

```python
# Verifica se há valores nulos no DataFrame
df.loc[df.isnull().any(axis=1)]

#Não há valores nulos no dataset
# df.isnull().sum() faz a mesma coisa realizando a soma de valores nulos por coluna
```

```python

# Sustitui os valores NaN por valores zeros (0s):
df.fillna(0)

# Sustitui os valores NaN pela média da coluna:
df.fillna(df.mean())

# Sustitui os valores NaN pela mediana da coluna:
df.fillna(df.median())
```

5) Compare os resultados obtidos usando um gráfico box plot. Que atributos possuem distribuições mais simétricas? Há a presença de valores extremos, denominados de outliers, para os atributos?

```python
for colum in df.columns[1:]:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='Class', y=colum, data=df)
    plt.title(f'Boxplot of {colum} by Class')
    plt.xlabel('Class')
    plt.ylabel(colum)
    plt.show()
```

6) No dataset escolhido selecione intervalos de amostragem (10%, 30% e 50% - Progressiva) e faça análises sobre os dados em relação as métricas estatísticas. O que você pode observar sobre os subconjuntos em relação as métricas estatísticas com a nova composição dos dados?

```python

sampling = [0.1, 0.3, 0.5, 0.7, 0.9, 1.0]

for s in sampling:
    print(f" \n{'-'*25} Sample size: {s*100}% {'-'*25}\n")
    n = int(len(df) * s)
    df_sample = df.sample(n=n, random_state=42)
    print(df_sample.describe())



```
